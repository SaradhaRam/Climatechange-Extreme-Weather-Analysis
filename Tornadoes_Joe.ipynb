```python
import os
import pandas as pd

tornado_raw = pd.DataFrame()
directory = "..\project_source_files_reduced"
for entry in os.scandir(directory):
        data = pd.read_csv(entry.path)
        tornado_raw = tornado_raw.append(data[['YEAR','EVENT_TYPE','TOR_F_SCALE']])
# display(tornado_raw)
tornado_cleaned = tornado_raw[(tornado_raw['EVENT_TYPE'] == 'Tornado') & ((tornado_raw["TOR_F_SCALE"] == "EF3") |
                                                                          (tornado_raw["TOR_F_SCALE"] == "EF4") |
                                                                          (tornado_raw["TOR_F_SCALE"] == "EF5"))]
tornado_count = pd.DataFrame(tornado_cleaned.groupby('YEAR')['EVENT_TYPE'].count())
display(tornado_count)
```
